# Part 1: Introduction to Async Python

## Async vs Threads vs Multiprocessing

| Feature | Asyncio | Threads | Multiprocessing |
|----------|----------|----------|----------|
| Best for | I/O-bound work | I/O-bound work | CPU-bound work |
| Parallelism | Cooperative | Concurrent | True parallelism |
| Overhead | Low | Medium | High |

### Example



````python
import asyncio

async def task():
    await asyncio.sleep(1)
    print("Done")
````

# Part 2: The Event Loop

The event loop pauses tasks when they are waiting and switches to other tasks.


````python
import asyncio

async def hello():
    print("Hello")
    await asyncio.sleep(1)
    print("World")

await hello()
````

# Part 3: async def and await



````python
import asyncio

async def greet(name):
    await asyncio.sleep(1)
    return f"Hello {name}"

result = await greet("Alice")
print(result)
````

# Part 4: Running Async Code

### Notebook

````python
await greet("Notebook User")
````

### Python Module

````python
import asyncio

async def main():
    print(await greet("Module User"))

if __name__ == "__main__":
    asyncio.run(main())
````

# Part 5: Async Gradio Callbacks


````python
import gradio as gr
import asyncio

async def slow_response(name):
    await asyncio.sleep(2)
    return f"Hello {name}"

gr.Interface(
    fn=slow_response,
    inputs="text",
    outputs="text"
).launch()
````

# Part 6: asyncio.gather


````python
import asyncio

async def task(name, delay):
    await asyncio.sleep(delay)
    return name

results = await asyncio.gather(
    task("A", 1),
    task("B", 2),
    task("C", 3),
)

print(results)
````

# Part 7: Tips and Traps

### Common mistakes

- Forgetting `await`
- Using `time.sleep()` inside async code
- Mixing blocking code with asyncio

Use:

```python
await asyncio.sleep(1)
```

instead of:

```python
import time
time.sleep(1)
```
